# Vector Operations

```python
y = np.array([0,1,0,1])
print(x,y)
```

```python
# Classical programming approach uses a lot of loops and it is slow
# Alternative 1
z = []
for i in range(0,x.size):
    z.append(x[i]*y[i])
z
```

```python
for a, b in zip(x,y):
  print(a,b)
```

```python
# Classical programming approach uses a lot of loops and it is slow
# Alternative 2
z = []
for a, b in zip(x,y):
  z.append(a*b)
z
```

```python
# Vectorized operations are fast and easy to write
z = x * y
z
```

```python
# Broadcasting: scalar values are applied to all elements of the array
print(x)
print(x*2)
print(x+1)
```

```python
# MEAN SQUARE of a vector 
ms = np.square(y).mean()
print(ms)
```
