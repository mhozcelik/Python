## MatPlotLib

```python
# MatPlotLib has a lot of options
plt.scatter(df_city['City_Area'],df_city['Population'])
```

```python
plt.plot(df_city['City_Area'],df_city['Population'],marker=".",linestyle="",color="red")
```

```python
plt.plot(df_city['City_Area'],df_city['Population'],marker=".",linestyle="",color="darkorange")
```

```python
# List of named colors
# https://matplotlib.org/stable/gallery/color/named_colors.html
```

```python
# SUBPLOTS
df_city00 = df_city.loc[df_city['City_Type']=='big metropol']
df_city01 = df_city.loc[df_city['City_Type']=='metropol']
df_city10 = df_city.loc[df_city['City_Type']=='urban']
df_city11 = df_city.loc[df_city['City_Type']=='rural']
```

```python
fig, ax = plt.subplots(2,2)
fig.suptitle('City Area vs Population')

ax[0,0].plot(df_city00['City_Area'],df_city00['Population'],marker=".",linestyle="",color="red")
ax[0,1].plot(df_city01['City_Area'],df_city01['Population'],marker=".",linestyle="",color="fuchsia")
ax[1,0].plot(df_city10['City_Area'],df_city10['Population'],marker=".",linestyle="",color="orange")
ax[1,1].plot(df_city11['City_Area'],df_city11['Population'],marker=".",linestyle="",color="blue")
```

```python
# HISTOGRAM
plt.hist(df_city["Population"],bins=20)
plt.xlabel("Population")
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.bar(df_city00["City_Name"],df_city00["City_Area"])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.barh(df_city10["City_Name"],df_city10["Population"])
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.pie(df_city00["Population"],labels=df_city00["City_Name"])
plt.show()
```
