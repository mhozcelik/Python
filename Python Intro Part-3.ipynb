# JOINING DATAFRAMES

```python
temp1 = df_city[['City_ID','City_Name']]
temp1.head(3)
```

```python
temp1.tail(3)
```

```python
temp = pd.DataFrame([[101,'unknown1']],columns=['City_ID','City_Name'])
temp
```

```python
temp1=temp1.append(temp)
```

```python
temp1.shape
```

```python
temp1.tail()
```

```python
temp2 = df_city[['City_ID','City_Area','Population']]
temp2.head(3)
```

```python
temp = pd.DataFrame([[102,1000,2000]],columns=['City_ID','City_Area','Population'])
temp
```

```python
temp2=temp2.append(temp)
```

```python
temp2.shape
```

```python
temp1.tail(4)
```

```python
temp2.tail(4)
```

```python
# merge is the most important function
pd.merge(temp1,temp2,on="City_ID").tail() 
```

```python
# "inner" is the default join type
pd.merge(temp1,temp2,on="City_ID",how="inner").shape 
```

```python
# "left" 
pd.merge(temp1,temp2,on="City_ID",how="left").tail()
```

```python
# "right" 
pd.merge(temp1,temp2,on="City_ID",how="right").tail()
```

```python
# "outer" 
temp3=pd.merge(temp1,temp2,on="City_ID",how="outer")
temp3.tail()
```

```python
pd.merge(temp1,temp2,how="inner",left_on='City_ID',right_on='City_ID').shape
```

```python
temp3['Pop_Density']=temp3['Population']/temp3['City_Area']
temp3.tail()
```
